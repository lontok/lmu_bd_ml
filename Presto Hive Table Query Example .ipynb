```python
import pandas as pd
from sqlalchemy import create_engine

from hdfs import InsecureClient
import os
```

```python
# create a connection to the glontok1 Hive database using the presto SQL engine
engine = create_engine('presto://glontok1@ec2-3-17-67-13.us-east-2.compute.amazonaws.com:8889/hive/glontok1')
```

```python
# assign SQL statement to a variable
sql = '''
SELECT *
FROM stores
'''
```

```python
# execute SQL and assign results to a Pandas DataFrame
df = pd.read_sql(sql, engine)
```
